```python
import seaborn as sns
import matplotlib.pyplot as plt  # Missing import added

# Load the car_crashes dataset
df_crashes = sns.load_dataset('car_crashes')

# Display the first few rows
print(df_crashes.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_crashes, x='speeding', y='total', hue='abbrev')
plt.title('Speeding vs Total Fatal Collisions')
plt.xlabel('Percentage of Speeding Drivers')
plt.ylabel('Total Fatal Collisions per Billion Miles')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
df_crashes_sorted = df_crashes.sort_values('ins_premium', ascending=False)
sns.barplot(data=df_crashes_sorted, x='abbrev', y='ins_premium', palette='viridis')
plt.title('Car Insurance Premiums by State')
plt.xlabel('State')
plt.ylabel('Insurance Premium ($)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = df_crashes.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Car Crashes Data')
plt.show()
```

```python
sns.pairplot(df_crashes, vars=['total', 'speeding', 'alcohol', 'ins_premium', 'ins_losses'])
plt.show()
```
